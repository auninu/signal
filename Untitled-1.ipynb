```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#karakteristik kuantisasi
bit=10
lama_waktu=1
tingkat_kuantisasi=1/1000
sinyal_delta=(bit/2)/((2**bit)-1) #dari rumus delta yang ada
t=np.arange(0, 1, tingkat_kuantisasi)#jumlah step kuantisasi
f=np.arange(0, 1000, 1) #frekuensi
```

```python
#input sinyal berupa gelombang sinusoidal 1 periode, saya pakai sin(2 pi omega t)
omega=10 #frekuensi dari sinyal input
sinyal_input=np.sin(2*np.pi*omega*np.exp(t**3))
```

```python
#fungsi untuk memodulasi
def modulasi(sinyal_input, sinyal_delta):
    sinyal_output=np.zeros(len(sinyal_input)) #membuat array berisi 0 di sepanjang sinyal input
    #Sinyal analog dibandingkan dengan sinyal delta, yaitu selisih antara nilai analog aktual dan nilai yang diprediksi sebelumnya.
    sinyal_prediksi=0
    for i in range(len(sinyal_input)):
        selisih_nilai=sinyal_input[i]-sinyal_prediksi
        #membandingkan, jika positif berarti 1 dan jika negatif berarti 0
        if selisih_nilai >=0:
            sinyal_output[i]=1
        else:
            sinyal_output[i]=0
        sinyal_prediksi = selisih_nilai*sinyal_delta #melakukan increment step dengan step size sebesar delta
    return sinyal_output
```

```python
#fungsi untuk membuat isyarat terkuantisasi
def kuantisasi(sinyal_input):
    sinyal_terkuantisasi=np.zeros(len(sinyal_input))
    for i in range(len(sinyal_input)):
        if sinyal_input[i] >=0:
            sinyal_terkuantisasi[i]=1
        else:
            sinyal_terkuantisasi[i]=0
       #sinyal_terkuantisasi[i]=sinyal_input[i]
    return sinyal_terkuantisasi
```

```python
#mengeplot sinyal

plt.figure(figsize=(10, 1))

#plot sinyal input
#plt.plot(t, sinyal_input)
plt.title("plot sinyal input")

#plot pulse train/bitstream
plt.plot(t, modulasi(sinyal_input, sinyal_delta))
#plt.title("plot sinyal terkuantisasi")

#plot isyarat terkuantisasi
plt.plot(t, kuantisasi(sinyal_input))

#plot isyarat input vs isyarat terkuantisasi
#plt.title("plot sinyal input vs sinyal terkuantisasi")

#mengeplot spektrum frekuensi
#plot frekuensi sinyal input
#X = np.fft.fft(sinyal_input)
#freq = np.fft.fftfreq(len(sinyal_input), tingkat_kuantisasi)
#plt.plot(freq, np.abs(X))
#plt.title("frekuensi sinyal input")

#plot frekuensi pulse train
#X = np.fft.fft(modulasi(sinyal_input, sinyal_delta))
#freq = np.fft.fftfreq(len(sinyal_input), tingkat_kuantisasi)
#plt.plot(freq, np.abs(X))
#plt.title("frekuensi pulse train")
#plot frekuensi isyarat terkuantisasi
#X = np.fft.fft(kuantisasi(sinyal_input))
#freq = np.fft.fftfreq(len(sinyal_input), tingkat_kuantisasi)
#plt.plot(freq, np.abs(X))
#plt.title("frekuensi isyarat terkuantisasi")



```
